#Zadanie 4. Utwórz tabele z innych tabel za pomocą DDL

2023-10-22 11_36_49-Ładowanie danych do BigQuery _ Zwiększenie umiejętności w Google Cloud.png

```python
%%bigquery

CREATE TABLE
  nyctaxi.january_trips AS
SELECT
  *
FROM
  nyctaxi.2018trips
WHERE
  EXTRACT(Month FROM pickup_datetime)=1; # 2018-05-04 05:39:26 UTC / pickup_datetime
```

2023-10-22 11_42_25-Ładowanie danych do BigQuery _ Zwiększenie umiejętności w Google Cloud.png

```python
%%bigquery

SELECT
  *
FROM
  nyctaxi.january_trips
ORDER BY
  trip_distance DESC
LIMIT
  1
```
